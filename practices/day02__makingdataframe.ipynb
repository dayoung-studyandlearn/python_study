```python
import pandas as pd
```

```python
# 데이터를 직접 입력해 데이터 프레임을 만들어보자
# 데이터 프레임을 만들 때는 pandas의 DataFrame()을 이용한다.
```

```python
df=pd.DataFrame({'name' : ['KimGijun', 'LeeJin', 'ParkDong', 'KimMingi'],
                 'english' : [90,80,60,70],
                 'math' : [50,60,100,20]})
df
# 인덱스는 0부터 시작한다.
# 함수 안에 중괄호 안에 딕셔너리 (key,  value(리스트도 가능))로 사용가능
```

```python
sum(df['english'])
# 시험 성적 중 영어 성적의 합
```

```python
sum(df['math'])/4
# 시험 성적 중 수학 성적의 평균
```

```python
# 혼자서 해보기
df_fruit=pd.DataFrame({'제품': ['사과', '딸기', '수박'],
                      '가격' : [1800, 1500, 3000],
                      '판매량' : [24, 38, 13]})
df_fruit
```

```python
sales_avg=sum(df_fruit['판매량'])/3
sales_avg
```

```python
price_avg=sum(df_fruit['가격'])/3
price_avg
```
