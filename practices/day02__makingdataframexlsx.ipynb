```python
# 여기서 전체 학생수를 잘 모르겠으면 len() 함수를 이용해보자
# len() 함수는 해당 리스트 값의 개수를 구하는 기능
df=pd.DataFrame({'a' : [1,2,2,3,4,4],
                'b' : [7,6,6,6,54,6],
                'c' : [2,1,9,7,4,7]})
df
```

```python
len(df)
# 둘 다 같은 값을 내뱉네. 어차피 갯수는 똑같다 이거겠지
```
