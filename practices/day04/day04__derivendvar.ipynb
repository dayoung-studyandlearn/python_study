```python
df_raw=pd.DataFrame({'var1':[2,3,1],
                    'var2':[5,4,2]})
df_raw
```

```python
# 파생변수 : 기존 변수를 변형해서 새로 만든 변수

df_var=pd.DataFrame({'var1':[2,5,3,6],
                    'var2':[5,6,1,2]})
df_var
```

```python
df_var['var_sum']=df_var['var1']+df_var['var2']
df_var
```

```python
df_var['var_mean']=df_var['var_sum']/2
df_var
```
