```python
import pandas as pd
exam=pd.read_csv('../../datasets/public/exam.csv')
exam
```

```python
# 오름차순으로 정렬하기
exam.sort_values('math') # math를 기준으로 오름차순 정렬
```

```python
exam.sort_values('math',ascending=False) # math를 기준으로 내림차순 정렬
```

```python
# 먼저오는 변수가 기준이 되고 ascending 없으면 오름차순이 기본값
exam.sort_values(['nclass','math'])
```

```python
# 여러개를 정렬기준으로 삼기도 가능
# nclass는 오름차순, science는 내림차순
exam.sort_values(['nclass','science'],ascending=[True,False])
```
