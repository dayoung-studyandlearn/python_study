```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../../datasets/public/midwest.csv')
df
```

```python
df=df.rename(columns={'poptotal':'total'})
```

```python
df=df.rename(columns={'popasian':'asian'})
df
```

```python
df['total'], df['asian']
```

```python
df['asianPercentage']=df['asian']/df['total']
```

```python
df['asianPercentage']=df['asianPercentage']*100
```

```python
df['asianPercentage']
```

```python
df['asianPer']=df['asian']/df['total']*100
df['asianPer']
```

```python
# 분포도 살펴보기 .hist()  히스토그램
df['asianPer'].plot.hist()
```

```python
df_mean=df['asianPer'].mean()
df_mean
```

```python
df['howasian']=np.where(df['asianPer']>=df_mean,'large','small')
df
```

```python
df['howasian']
```

```python
df['howasian'].value_counts().plot.bar()
```
